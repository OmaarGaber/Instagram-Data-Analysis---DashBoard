```python
import pandas as pd

def g(df):
    for i in df.index:
        i = int (i)
        if type(df.loc[i, 'business_category_name']) == float:
            if df['business_category_name'].groupby(df['followers']<20000) and df['business_category_name'].groupby(df['followers']>15000) :
                df.loc[i, 'business_category_name'] = 'Non-Profits & Religious Organizations'
                
                
            elif df['business_category_name'].groupby(df['followers']<50000 and df['business_category_name'].groupby(df['followers']>30000)):
                df.loc[i, 'business_category_name'] = 'Creators & Celebrities'
                
            elif df['business_category_name'][i].loc(df['followers']>15000):
                df[df['business_category_name'][i]] = 'Business & Utility Services'
                

            elif df['business_category_name'].loc(df['followers']<5000):
                df[df['business_category_name'][i]] = 'Personal Account'
            

    return df

df = g(df.copy())
# print(df)
```

### Removing Outlaiers

```python
plt.boxplot(df['followers'])
```

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range , upper_range
```

```python
lowincome,uppincome=remove_outlier(df['followers'])
df['followers']=np.where(df['followers']>uppincome,uppincome,df['followers'])

df['followers']=np.where(df['followers']<lowincome,lowincome,df['followers'])
```
